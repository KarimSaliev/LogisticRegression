```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import RandomOverSampler
```

```python
df = pd.read_csv('breast-cancer.csv', index_col=0)
df
```

```python
df.head(12) #first 12 rows
```

```python
df.tail(5) #last 5 rows
```

```python
df.shape #no of rows and cols
```

```python
df.isna().sum() # dataset doesnt have any missing values in its columns
```

```python
label_encoder = LabelEncoder()
df['diagnosis'] = label_encoder.fit_transform(df['diagnosis'])
```

```python
X = df.drop('diagnosis', axis=1)
Y = df[['diagnosis']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.25, random_state=42)
```

```python
print(len(y_train[y_train['diagnosis']==1]))
print(len(y_train[y_train['diagnosis']==0]))
```

```python
#After determining that B class instances dominate the dataset, I make the decision to oversample the data
#to ensure unbiased results.
```

```python

ros = RandomOverSampler()
x_train,y_train = ros.fit_resample(X,Y)
```

```python
print(len(y_train[y_train['diagnosis']==1]))
print(len(y_train[y_train['diagnosis']==0]))
#After balancing the data, we can proceed to train the model
```

```python
lr_model = LogisticRegression(max_iter=3000) #increasing number of iterations 
```

```python
y_train = y_train.values.ravel() #converting 2d dataframe to 1d array to match model requirements
```

```python
lr_model.fit(x_train,y_train)
```

```python
y_pred = lr_model.predict(x_test)
```

```python
print(y_test.values.ravel()[:10]) # actual data
print(y_pred[:10]) #predicted data
```

```python
cm = confusion_matrix(y_test,y_pred)
acc = accuracy_score(y_test, y_pred)
clf_report = classification_report(y_test,y_pred)
print("Confusion matrix:")
print(cm)
print("Accuracy score:", acc)
print("Classification report:\n", clf_report)
```
